```python
import pandas as pd

# convert the "Creation Date" column to a datetime object and extract the date component
clean_mdf2['Creation Date'] = pd.to_datetime(clean_mdf2['Creation Date']).dt.date

# group the dataset by the "Creation Date" and "Neighborhood Council" columns and get the count of rows for each combination
cmdf_bydate_nc2 = clean_mdf2.groupby(['Creation Date', 'Neighborhood Council']).size().reset_index(name='Total Rows')

# print the resulting grouped DataFrame
print(cmdf_bydate_nc2)
```

```python
# convert the "Creation Date" column to a datetime object and extract the date component
clean_mdf['Creation Date'] = pd.to_datetime(clean_mdf['Creation Date']).dt.date

# group the dataset by the "Creation Date" and "Neighborhood Council" columns and get the count of rows for each date and council
cmdf_bydate_nc = clean_mdf.groupby(['Creation Date', 'Neighborhood Council']).size().reset_index(name='Total Rows')

# remove rows with missing values in the "Neighborhood Council" column
cmdf_bydate_nc = cmdf_bydate_nc.dropna(subset=['Neighborhood Council'])

# print the resulting grouped DataFrame
print(cmdf_bydate_nc)
```
